# Load package

```python
import matplotlib.pyplot as plt
```

# Work with lists

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
```

# Plot a line

```python
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

# Plot multiple series

```python
x2 = [2, 4, 6]
y2 = [8, 64, 216]
```

```python
fig, ax = plt.subplots( nrows=2 )
ax[0].plot( x, y, '.' )
ax[1].plot( x2, y2 )
plt.show()
```

# Annotate plot

```python
fig, ax = plt.subplots()
ax.plot( x, y, label="x^2" )
ax.plot( x2, y2, '+', label="x^3" )
ax.set_title( "Polynomials" )
ax.set_xlabel( "x" )
ax.set_ylabel( "f(x)" )
ax.legend()
plt.show()
```

# Subplot

```python
fig, ax = plt.subplots( nrows=2 )
ax[0].plot( x, y, '.' )
ax[1].plot( x2, y2 )
plt.show()
```

# Question 1 - Change axes on subplot

```python
fig, ax = plt.subplots( nrows=2, sharex = True, sharey = True )
ax[0].plot( x, y, '.' )
ax[0].set_title( "Squares" )
ax[0].set_xlabel( "x" )
ax[0].set_ylabel( "x^2" )
ax[1].plot( x2, y2 )
ax[1].set_title( "Cubes" )
ax[1].set_xlabel( "x" )
ax[1].set_ylabel( "x^3" )
plt.show()
```
